Pandas is mainly for data manipulation and analysis since it has efficient data structures. The quick highlights would be 
- DataFrame - an efficient data struture to handle large data.
- Reshaping and pivoting of data is easy and fast
- Slicing, indexing and subsetting of large data sets
- Efficient merging and joining of data
- Efficinet aggregations

Pandas has three speciliazed data type
- Series
- DataFrame
- Panel

## Series

Single dimensional <br>
Size cannot be change once created <br>
Values can be altered <br>
To create we invoke - **pandas.Series(data, index, dtype)** <br>

```python
# a sample of a series
import pandas as pd
import numpy as np
data = np.array(['a','b','c','d'])
s = pd.Series(data)
print(s)
```

## DataFrame

2 dimensional <br>
Can contain heterogenous data i.e. different data types <br>
Size and values can be changed <br>
To invoke we use - **pandas.DataFrame( data, index, columns, dtype)**

```python
import pandas as pd
data = [1,2,3,4,5]
df = pd.DataFrame(data)
print(df)
```

```python
import pandas as pd
data = [['Alex',10],['Bob',12],['Clarke',13]]
df = pd.DataFrame(data,columns=['Name','Age'])
print(df)
```

```python
import pandas as pd
data = {'Name':['Tom', 'Jack', 'Steve', 'Ricky'],'Age':[28,34,29,42]}
df = pd.DataFrame(data)
print(df)
```

Lets have a look at the data load function. The most common one being
pandas.read_csv(file_name,sep=",",header=infer,names,dtype,na_values)

header is the values to be used as header for the document. If no values exist in the document you can use None. By default it infers the header form first row <br>
name can be used to pass a list of column names <br>
na_values is list of values to be considered as NA <br>

pandas.read_excel(), pandas.read_json() do exist. Ref [__Pandas Docs__](https://pandas.pydata.org/pandas-docs/stable/io.html)
